```python
import pandas as pd
import numpy as np
```

```python
col_name = ['user_id','item_id','rating','timestamp']
data = pd.read_csv('userdata.data',sep='\t',names=col_name)
data.head()
```

```python
movies = pd.read_csv('movie_title')
movies.head()
```

```python
df = pd.merge(movies,data,on='item_id')
df.head()
```

```python
dk = pd.DataFrame(df.groupby('title')['rating'].mean())
dk.head()
```

```python
dk['num_rating'] = pd.DataFrame(df.groupby('title')['rating'].count())
dk.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


dk.num_rating.hist(bins=70)
```

```python
sns.jointplot('rating','num_rating',dk)
```

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
movie_rating = moviemat['Star Wars (1977)']
movie_rating.head()
```

```python
similar = pd.DataFrame(moviemat.corrwith(movie_rating),columns=['correlation'])
similar.dropna(inplace=True)
```

```python
similar.sort_values('correlation',ascending=False).head()
```

```python
similar = similar.join(dk['num_rating'])
similar.head()
```

```python
similar[similar['num_rating']>100].sort_values('correlation',ascending=False).head()
```
